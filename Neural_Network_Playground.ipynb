# Neural Network Playground

```
import tensorflow 
from tensorflow import keras
from tensorflow.keras import *
from tensorflow.keras.layers import *

import numpy as np
import matplotlib.pyplot as plt
```

# Prepare Data

```
outer = 0.5
x = np.linspace(-outer, outer, 1000)
y = np.sqrt(outer**2 - x**2)
x = np.append(x, x)
y = np.append(y, -y)
label = np.ones(len(x))

plt.figure(figsize=(5,5))
plt.scatter(x, y)
```

```
x0 = x
y0 = y
label0 = label
```

```
inner = 0.2
x = np.linspace(-inner, inner, 1000)
y = np.sqrt(inner**2 - x**2)
x = np.append(x, x)
y = np.append(y, -y)
label = np.zeros(len(x))

x = np.append(x, x0)
y = np.append(y, y0)
label = np.append(label, label0)

plt.figure(figsize=(5,5))
plt.scatter(x, y, c=label)
```

# Train

```
import tensorflow.keras.optimizers

model = Sequential([
    Dense(3, activation='relu', input_shape=(2,), use_bias=False),
    Dense(1, activation='relu', use_bias=False),
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)
```

```
data = np.column_stack([x, y])

model.fit(data, label, epochs=400)
```
